```python
import os, ast, sqlite3
import pandas as pd
```

```python
os.listdir('../')
```

```python
conn = sqlite3.connect('../db.sqlite3')
cursor = conn.cursor()
```

```python
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = [row[0] for row in cursor.fetchall()]
tables
```

```python
slaves = pd.read_sql_query(f'SELECT * FROM Devices_device WHERE master=false', conn).esp_id.tolist()
slaves
```

```python
query = f"SELECT * FROM Logs_log WHERE esp_id IN ({', '.join(['?'] * len(slaves))}) ORDER BY id DESC LIMIT 15"
print(query)
```

```python
df = pd.read_sql_query(query, conn, params=slaves)
df
```

```python
df['data'] = df.data.apply(ast.literal_eval)
df
```

```python
temps = [data.get('temperature') for data in df.data if data.get('temperature') is not None]
temps
```
